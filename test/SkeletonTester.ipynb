# SkeletonTester

## Imports

```python
%matplotlib inline

# Imports
import numpy as np
import os as os
import matplotlib.pyplot as plt
```

## Reading output files

```python
durations = np.array([])

m_vertices = np.array([])
m_edges = np.array([])

s_vertices = np.array([])
s_edges = np.array([])

for path in os.listdir():
    if(path.endswith(".out")):
        file = open(path, 'r')
        lines = file.readlines()
        
        # Get durations
        durations = np.append(durations, int(lines[0][:-1]))
        
        # Get starting and ending number of vertices
        vertices = lines[1][:-1].split(" ", 1)
        edges = lines[2][:-1].split(" ", 1)
        
        m_vertices = np.append(m_vertices, int(vertices[0]))
        m_edges = np.append(m_edges, int(edges[0]))   
        
        s_vertices = np.append(s_vertices, int(vertices[1]))
        s_edges = np.append(s_edges, int(edges[1]))
```

## Durations in fuction of starting vertices and edges

```python
## Starting vertices
plt.scatter(m_vertices, durations)
plt.show()
```

```python
## Starting edges
plt.scatter(m_edges, durations)
plt.show()
```

## Ending vertices and edges in fuction of starting vertices and edges

```python
## Starting vertices
plt.scatter(m_vertices, s_vertices)
plt.show()
```

```python
## Starting edges
plt.scatter(m_edges, s_edges)
plt.show()
```
